```python
import numpy as np
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt
import pandas as pd
import math
import xlrd 
from random import seed
from random import random
import scipy.fftpack 

seed(1)
I = 1j
D = 0.2 #Aperture size- 20 cm 
r0 = 0.1345 #Fried parameter proportional lambda^(6/5), Lambda = 640nm 

nx = 100
ny = 100
dx = 5
dy = 5
unitflag = 0
decenter_x = 0
decenter_y = 0

details=np.zeros((1,7))
details[0][0]=nx
details[0][1]=ny
details[0][2]=dx
details[0][3]=dy
details[0][4]=unitflag
details[0][5]=decenter_x
details[0][6]=decenter_y

fi = open('imag'+ str(3)+'.DAT','ab')
fr = open('real'+str(3)+'.DAT', 'ab')
fi.truncate(0)
fr.truncate(0)
np.savetxt(fr,details)
np.savetxt(fi,details)

def Fconv(k,l):
    v = math.sqrt(k**2+l**2)
    psi = -math.pi + 2*math.pi*random()
    return math.sqrt(0.023)*((2*D/r0)**(5/6)) * (v**(-11/6))* np.exp(I*psi) 

arr = [ [0]*100 for i in range(100)]
realmatrix = np.zeros((10000, 4))
imagmatrix = np.zeros((10000, 4))
#real = [0]*10000
#imag = [0]*10000


for k in range (1,nx):
        for l in range (1,ny):
            arr[k-1][l-1] = Fconv(k,l)
            
screen = scipy.fftpack.fft2(arr)

k=0
for i in range(nx):
    for j in range(ny):
        realmatrix[k][0]=screen[i][j].real
        imagmatrix[k][0]=screen[i][j].imag
        k=k+1
        
np.savetxt(fr,realmatrix)
np.savetxt(fi,imagmatrix)
#df = pd.DataFrame(realmatrix)
#df.to_excel(excel_writer = "C:/RupaRocz/Research and projects/Master thesis/phasescreen1_real.xlsx")
#df1 = pd.DataFrame(imagmatrix)
#df1.to_excel(excel_writer = "C:/RupaRocz/Research and projects/Master thesis/phasescreen1_imag.xlsx")
fr.close()
fi.close()
```
